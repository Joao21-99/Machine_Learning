### 2. Data preparation

```python
# drop missing value
df = df.dropna()
df.info()
```

```python
#drop object columns
df = df.select_dtypes(exclude=['object'])
df
```

```python
#Define X and y for train
X = df.drop('Survived', axis= 1)
y = df['Survived']
```

```python
#shape of X and y
X.shape, y.shape
```

```python
# import and train the model
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
```

```python
#score the model
neigh.score(X,y)
```
